````python
def payoffCall(st, pr, strike):
    hDistribution = np.max([st-strike, np.zeros(len(st))],axis = 0)
    h = hDistribution@pr
    
    plt.plot(st, hDistribution)
    plt.title("Option payoff per stock price")
    plt.xlabel('Stock price')
    plt.ylabel('Option payoff')
    plt.show()
    return(h)
````

## Problem 4: The lognormal stock price distribution

In this exercise, we compute the stock price density at maturity $T$ , i.e. the density of $S_T$ , using three different models. We assume that the initial price $S_0 = 1$, the drift $\mu = r = 0.05$, the volatility $\sigma = 0.2$, and the maturity $T = 0.5$ years.

(a) The stock price follows the geometric Brownian motion, so that

$$S_T = S_0 +\int_0^T
rS_sds +
\int_0^T
\sigma S_sdX_s.$$

We partition the time interval $[0, T]$ to n discrete points with length $dt = T/n$, so that $0 = t_1 < t_2 <\dots < t_n=T$. Let us denote the stock price at time $t_i$ by $s_i$, $1 \leq i\leq n$. The Brownian motion $X_{t_{i+1}} − X_{t_i}$ is modelled as $z\sqrt{dt}$, where $z\sim\mathcal{N}(0, 1)$. In `Python`, a hundred normally distributed random variables with mean 0 and variance 1 are obtained with the commands `np.random.normal(0, 1)`. You can search `Python` documentation help to get more information of these commands.

````python
r     = 0.05 # Risk-free interest rate
sigma = 0.2 # Standard deviation
T     = 0.5 # Time to maturity
n     = 100 # Number of steps
dt    = T/n # Incremental time
stock = np.ones(n); # Stock price vector, initial value 1
np.random.seed(123)
````

We are then ready to compute one realization of the stock price process, i.e., the discrete stock prices $s_1, s_2, \dots , s_n$. We set the initial stock price $s(1) = 1$ and use a loop

````python
prev = 1
for i in range(1, n):
  prev = stock[i-1]
  stock[i] = prev + r * prev * dt + sigma * prev * np.sqrt(dt) * np.random.normal(0, 1)
````

````python
N = 1000
multi_stock = np.ones([n,N]);
````

and use two loops:

````python
for j in range(N):
    for i in range(1, n):
        multi_stock[i, j]= multi_stock[i - 1, j] + r * multi_stock[i - 1, j] * dt + sigma * multi_stock[i - 1, j] * np.sqrt(dt) * np.random.normal(0, 1)
````

Vectorization (and choosing the random variables outside the loop) will make this much faster

````python
multi_stock = np.ones([n, N])
sqrt_dt = np.sqrt(dt)
random_nos = np.random.normal(0, 1, [n - 1, N])
for i in range(1, n):
    multi_stock[i, :] = multi_stock[i - 1, :] * (1 + r * dt + sigma * sqrt_dt * random_nos[i - 1, :])
````

To plot the N stock price processes, you can write:

````python
plt.plot(multi_stock)
plt.title("Random stock price processes")
plt.xlabel("t")
plt.ylabel("Stock price")
plt.show()
````

You can also plot the final values $S_T$ in a histogram with approx. 50 discrete points:

(b) The stock price $S_T$ is given in exponential form by

$$S_T = S_0 \exp((r − \frac{1}{2}\sigma^2 )T + \sigma W_T ).$$

Remember from last week and Exercise 3 that we can write this as

$$S_T = S_0 \exp((r−\frac{1}{2}\sigma^2 )T + \sigma\sqrt{T}Z),$$

where $Z$ is the normal random variable, $Z\sim\mathcal{N}(0, 1)$. Now, given the initial stock price $S_0 = 1$, as well as $r$, $\sigma$ and $T$ , we can directly compute a possible price at maturity, $S_T$. As above, we compute $N$ prices:

````python
multi_stock_direct = np.ones(N)
z = np.random.normal(0, 1, N)
S0 = 1
#
multi_stock_direct = S0 * np.exp((r - 0.5 * sigma**2) * T + sigma * np.sqrt(T) * z);
````

The corresponding histogram is obtained by

(c) As we know that the distribution of $S_T$ is lognormal, $S_T\sim\mathcal{lN}(\log(S_0) + (r -\frac{1}{2}\sigma^2)T, \sigma^2 T)$, we don't need to compute $N$ values for $S_T$. Instead, we can use the `Python` function `lognorm.pdf`, presented in problem (3).

The support for the distribution of $S_T$ can be chosen as

````python
nn = 100
S0 = 1
tt = np.linspace(0.0001, 2, nn)
````

The possible values of $S_T$ are then the n discrete points t. The probability of each of these points (probability of each price $S_T$) is obtained with the lognormal distribution `PrST`:

````python
PrST = lognorm.pdf(tt, sigma * np.sqrt(T), scale = S0 * np.exp((r - .5 * sigma**2) * T))
````

You can plot the lognormal distribution:

````python
plt.plot(tt, PrST)
plt.xlabel("Stock price at maturity")
plt.ylabel("Density")
plt.title("Probability distribution of stock price at maturity")
plt.show()
````

Note the different notation in `Python` and the lecture slides. For a lognormally distributed random variable, $Y\sim\mathcal{lN}(\mu, \sigma^2)$, we write in `Python`

````python
Y = lognorm.pdf(tt, sigma, scale=np.exp(mu));
````

using the standard deviation σ, not the variance σ².

You can now compare the three different price distributions of $S_T$. Notice that the histograms in (4a) and (4b) are not scaled to illustrate probabilities (total probability must be one). You can try different values for mu, sigma, T and N. Remember to define E.

Then, based on the lognormal distribution St in (4c), compute the payoff of a call with strike E and maturity T. The payoff at each support point $t_i$ is given by $h_i = \max(t_i-E, 0)$, and the corresponding probability is $PrST (t_i) = PrST (h_i)$. The expectation of the call is:
$$
\mathbb{E}[C]=\frac{M}{n}\sum_{i=1}^n PrST(h_i)*h_i
$$
where $M = \max(t) −\min(t)$. As the support t and the probability function $PrSt$ are row vectors of size $(1 \times n)$, the same result is obtained directly with
$$
\mathbb{E}[C]=\frac{M}{n}h*PrST^T
$$

````python
import QuantLib as ql
````

````python
# Parameters
S = 50.0
E = np.array([30, 35, 40, 45, 55, 60]) * 1.0 # needs to be of type float 
T = 0.5
sigma = 0.5
r = 0.06
````

````python
# Settings
day_count = ql.Actual365Fixed()
calendar = ql.NullCalendar()
calculation_date = ql.Date().todaysDate()
ql.Settings.instance().evaluationDate = calculation_date
````

````python
# European Options Pricing Function
def european_option_price(option_type, S0, K, T, r, sigma):
    payoff = ql.PlainVanillaPayoff(option_type, K)
    exercise = ql.EuropeanExercise(calculation_date + int(T * 365))
    option = ql.VanillaOption(payoff, exercise)

    spot_handle = ql.QuoteHandle(ql.SimpleQuote(S0))
    volatility_handle = ql.BlackVolTermStructureHandle(
        ql.BlackConstantVol(calculation_date, calendar, sigma, day_count)
    )
    risk_free_handle = ql.YieldTermStructureHandle(
        ql.FlatForward(calculation_date, ql.QuoteHandle(ql.SimpleQuote(r)), day_count)
    )

    bsm_process = ql.BlackScholesProcess(spot_handle, risk_free_handle, volatility_handle)
    option.setPricingEngine(ql.AnalyticEuropeanEngine(bsm_process))

    return option.NPV()
````

````python
# Calculate prices
call_prices = [european_option_price(ql.Option.Call, S, K, T, r, sigma) for K in E]
put_prices = [european_option_price(ql.Option.Put, S, K, T, r, sigma) for K in E]

print("Calls: ", call_prices)
print("Puts: ", put_prices)
````

First, try to write the code yourself. Below you have the correct code. We have added machine epsilon eps = 10⁻¹⁶ in the denominator of d₁ to prevent division by 0. Python can get confused when dividing something by 0.

````python
S = 30.0
E = 30.0
Cobs = 2.5
r = .06
T = 0.25
````

We use Newton's method to recover the implied volatility. Here the machine $\epsilon$ term ```eps``` is added to avoid division by zero.

````python
def vola(S, E, r, T, Cobs):
    vola = .5
    d = 1
    while(d > eps):
        d1 = (np.log(S / E) + (r + vola ** 2 / 2) * T) / (vola * np.sqrt(T) + eps)
        d2 = d1 - vola * np.sqrt(T)
        C = S * norm.cdf(d1) - E * np.exp(-r * T) * norm.cdf(d2)
        vDer = S * np.sqrt(T) * np.exp(-d1 ** 2 / 2) / (np.sqrt(2 * np.pi) + eps)
        d = (Cobs - C) / -vDer
        vola = vola - d
    return vola

print(vola(S, E, r, T, Cobs))

op1 = BSM(kind='call', S0=S, K=E, T=T, r=r, sigma=sigma)
print(op1.implied_vol(value=Cobs))
````

You can also use a more general form for several options and strikes:

````python
E    = np.array([22, 24, 26, 28, 30, 32, 34, 36, 38])
Cobs = np.array(list(map(lambda k: BSM(kind='call', S0=S, K=k, T=T, r=r, sigma=0.1).value(), E )))
````

where ```L``` is the amount of options with different strikes

````python
def vola_m(S, E, r, T, Cobs):
    if(len(E) != len(Cobs)): return None
    L = len(E)
    vola = .5 * np.ones(L)
   
    for k in range(L):
        d = 1
        if(np.isnan(Cobs[k])): 
            vola[k] = np.nan
            d = 0
        while(d > eps):
            d1 = (np.log(S / E[k]) + (r + vola[k] ** 2 / 2) * T) / (vola[k] * np.sqrt(T) + eps)
            d2 = d1 - vola[k] * np.sqrt(T)
            C = S * norm.cdf(d1) - E[k] * np.exp(-r * T) * norm.cdf(d2)
            vDer = S * np.sqrt(T) * np.exp(-d1 ** 2 / 2) / np.sqrt(2 * np.pi) + eps
            d = (Cobs[k] - C) / -vDer
            vola[k] = vola[k] - d
            
    return vola

print(vola_m(S, E, r, T, Cobs))
````

We can use the defined functions from Problem 1, we'll use the ready function from `pyfinance`, which does basically the same thing as our own function `vola`, but faster (as it is written in C++). We will add a new column to our data frame, which contains the volatilities. We should notice, that there is not always an implied (Black-Scholes) volatility for all parameter combinations, for these will set `NaN`.

````python
import numpy as np
import pandas as pd
from scipy.optimize import brentq
from scipy.stats import norm

# Black–Scholes price for European options (with dividend yield q)
def bs_price(S, K, T, r, sigma, is_call=True, q=0.0):
    S = float(S); K = float(K); T = float(T); r = float(r); sigma = float(sigma); q = float(q)
    if T <= 0 or S <= 0 or K <= 0:
        # At expiry or invalid inputs: return intrinsic value
        if is_call:
            return max(0.0, S - K)
        else:
            return max(0.0, K - S)
    sqrtT = np.sqrt(T)
    d1 = (np.log(S / K) + (r - q + 0.5 * sigma * sigma) * T) / (sigma * sqrtT)
    d2 = d1 - sigma * sqrtT
    if is_call:
        return S * np.exp(-q * T) * norm.cdf(d1) - K * np.exp(-r * T) * norm.cdf(d2)
    else:
        return K * np.exp(-r * T) * norm.cdf(-d2) - S * np.exp(-q * T) * norm.cdf(-d1)

def no_arb_bounds_call(S, K, T, r, q=0.0):
    # European call bounds
    lb = max(0.0, S * np.exp(-q * T) - K * np.exp(-r * T))
    ub = S * np.exp(-q * T)
    return lb, ub
````

````python

def implied_vol_call(price, S, K, T, r, q=0.0, tol=1e-8, max_iter=100):
    # Return NaN if inputs are invalid
    if any(map(lambda x: x is None or np.isnan(x) or np.isinf(x), [price, S, K, T, r, q])):
        return np.nan
    if T <= 0 or S <= 0 or K <= 0 or price < 0:
        return np.nan

    lb, ub = no_arb_bounds_call(S, K, T, r, q)
    # If price violates bounds, no solution
    if price < lb - 1e-10 or price > ub + 1e-10:
        return np.nan
    # If price is essentially at the lower bound, vol ~ 0
    if abs(price - lb) <= 1e-12:
        return 0.0

    # Objective function for root finding
    def f(sig):
        return bs_price(S, K, T, r, sig, is_call=True, q=q) - price

    # Try to bracket the root; start small sigma and grow upper bound until sign change
    a = 1e-9
    b_candidates = [0.5, 1.0, 2.0, 3.0, 5.0, 10.0, 20.0, 50.0]
    fa = f(a)
    root = np.nan
    for b in b_candidates:
        fb = f(b)
        if fa * fb < 0:
            # Root bracketed, solve with Brent
            try:
                root = brentq(f, a, b, xtol=tol, rtol=tol, maxiter=max_iter)
            except ValueError:
                root = np.nan
            break
        fa = fb
        a = b

    return root
````

````python

# Example: compute implied vol for each row of your DataFrame
# Assumes data_long has columns: 'S', 'E' (strike), 'T' (years), 'r', 'Cobs'
def compute_iv_series(df, q=0.0):
    S = df['S'].to_numpy()
    K = df['E'].to_numpy()
    T = df['T'].to_numpy()
    r = df['r'].to_numpy()
    P = df['Cobs'].to_numpy()

    iv = np.empty(len(df), dtype=float)
    for i, (p, s, k, t, rr) in enumerate(zip(P, S, K, T, r)):
        iv[i] = implied_vol_call(price=p, S=s, K=k, T=t, r=rr, q=q)
    return iv

# Usage:
data_long['volatility'] = compute_iv_series(data_long, q=0.0)
print(data_long)
````
