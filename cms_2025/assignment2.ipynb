### Exercise 1

As part of a software for modelling defects on fabricated silicon wafers you need to
simulate a uniform distribution inside a circle. (You can choose the radius to be ùëÖ = 1.) In
each of the cases below, generate 2000 points (ùë•, ùë¶) and plot them. For comparison, you
can also generate more points, for example 20 000 points; in doing so, you see why such
geometric distributions sometimes need to be checked with a small number of points.


a) (w = 2) First implement this using the rejection method by thinking of the circle being
enveloped by a square. Generate 2000 points (ùë•, ùë¶) and plot them.


```python
n_points = 2000
accepted_points = []

while len(accepted_points) < n_points:
    # points in the square
    x = np.random.uniform(-1, 1)
    y = np.random.uniform(-1, 1)

    # if point is in the circle (rejection sampling)
    if x**2 + y**2 <= 1:
        accepted_points.append((x, y))

accepted_points_np = np.array(accepted_points)


plt.figure(figsize=(6, 6))
plt.scatter(accepted_points_np[:, 0], accepted_points_np[:, 1], s=5)
plt.title('Distribution using Rejection Method')
plt.xlabel('x')
plt.ylabel('y')
plt.axis('equal')
plt.grid(True)
plt.show()
```

b) (w = 2) You want to make the simulation computationally more effective and not to
reject any points, so you generate random points inside the circle by using polar
coordinates, ùë• = ùëü cos(2ùúãùúÉ) , ùë¶ = ùëü sin(2ùúãùúÉ), and drawing ùëü and ùúÉ from uniform
distributions. Generate 2000 points (ùíô, ùíö) and plot them. Explain the outcome and
the reason for it.


```python
n_points = 2000

# generate theta from uniform distribution 
theta = np.random.uniform(0, 2 * np.pi, n_points)

# generate r from a uniform distribution 
r = np.random.uniform(0, 1, n_points)

# Convert to Cartesian coordinates
x = r * np.cos(theta)
y = r * np.sin(theta)


plt.figure(figsize=(6, 6))
plt.scatter(x, y, s=5)
plt.title('Distribution using Polar Coordinates')
plt.xlabel('x')
plt.ylabel('y')
plt.axis('equal')
plt.grid(True)
plt.show()
```

c) Simulation of the circular uniform distribution by the inverse distribution method:
Generate (ùë•, ùë¶) from polar coordinates again drawing ùúÉ from uniform distribution but
applying a proper transformation for the radius ùëü ‚àà (0, ùëÖ): It is straightforward to see
that the correct pdf is of the form ùëù(ùëü) = ùê∂2ùúãùëü, where ùê∂ is a constant to be
determined from the condition ùêπ(ùëü = ùëÖ) = 1. ùêπ is the cumulative distribution for ùëü.
(i) (w = 2) Determine the proper transformation for generating ùíì. Show the
derivation of this transformation in your answer.
(ii) (w = 2) Implement an algorithm and generate and plot 2000 points (ùíô, ùíö)
using this method.

**Derivation**:

```python
n_points = 2000

# generate theta from a uniform distribution 
theta = np.random.uniform(0, 2 * np.pi, n_points)

# generate u from a uniform distribution 
u = np.random.uniform(0, 1, n_points)

# using the inverse transform to get r
r = np.sqrt(u)

# Convert to Cartesian coordinates
x_inverse = r * np.cos(theta)
y_inverse = r * np.sin(theta)

plt.figure(figsize=(6, 6))
plt.scatter(x_inverse, y_inverse, s=5)
plt.title('Distribution using Inverse Transform Method')
plt.xlabel('x')
plt.ylabel('y')
plt.axis('equal')
plt.grid(True)
plt.show()
```

d) (w = 1) Compare the outcomes of b) and c). Explain the difference. (Hint: See the
explanation of the method and the examples after that in Lecture 2.)

**Answer**:

The plot from the polar coordinate method clearly shows a non-uniform distribution, with a high concentration of points near the center and a lower density towards the edge of the circle.
While the plot from the inverse distribution method, shows a distribution that seems uniform across the entire circle, just like the result from the rejection method.

In the polar coordinate method, sampling r from a uniform distribution gives equal probability to any interval of r. However, as we go away from the centre, we need more points to cover the area of the inner circle. This leads to non-uniform spatial distribution. For example, the area of the inner circle between r and r + dr is approximately $2 \pi r dr$. This means the probability of a point falling in the inner circle should be proportional to r.
While in inverse transform method, the transformation r = $\sqrt u$, where u is uniform distribution, we are effectively sampling r from a distribution with the pdf p(r) = 2r. This ensures that the number of points in any given area of the circle is proportional to that area, resulting in a truly uniform spatial distribution of points.

b) (w = 2) Do linear binning to view the pdf of the generated data. Plot the binned data
as a histogram of the bins in linear coordinates. Then plot the binned data as a
histogram of the bins in logarithmic coordinates. In all plots include also the line ùíô!ùüë ùüê 
possibly multiplied by an appropriate coefficient for the line to ‚Äúsweep‚Äù close to

```python

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

#  Linear coordinates plot 
# histogram
counts, bin_edges = np.histogram(x, bins=30)
bin_widths = np.diff(bin_edges)
bin_midpoints = bin_edges[:-1] + bin_widths / 2

# Plot
ax1.hist(x, bins=30, alpha=0.7, label='Linearly Binned Data')
ax1.set_title('Linear Binning in Linear Coordinates')
ax1.set_xlabel('x')
ax1.set_ylabel('Frequency')

#  reference line 
scaling_factor_linear = 1e4
x_vals_linear = np.linspace(min(x), max(x), 100)
y_vals_linear = scaling_factor_linear * x_vals_linear**(-3/2)
ax1.plot(x_vals_linear, y_vals_linear, 'r-', label='Cx^(-3/2)')
ax1.legend()
ax1.grid(True)
ax1.set_ylim([0, 900])


# Logarithmic coordinates plot 
# histogram in log-log scale
ax2.hist(x, bins=30, alpha=0.7, label='Linearly Binned Data')
ax2.set_title('Linear Binning in Logarithmic Coordinates')
ax2.set_xlabel('x')
ax2.set_ylabel('Frequency')
ax2.set_xscale('log')
ax2.set_yscale('log')


# reference line 
x_vals_log = np.logspace(np.log10(min(x)), np.log10(max(x)), 100)
scaling_factor_linear = 1e4
y_vals_log = scaling_factor_linear * x_vals_log**(-3/2)
ax2.plot(x_vals_log, y_vals_log, 'r-', label='Cx^(-3/2)')
ax2.legend()
ax2.grid(True, which="both", ls="--")
ax2.set_ylim([1, 1000])


plt.tight_layout()
plt.show()
```

d) (w = 2) Do logarithmic binning of the pdf. Plot the binned data as midpoints of the
bins in logarithmic coordinates including the line ùíô!ùüë ùüê

```python
# logarithmically spaced bin edges
log_bins = np.logspace(np.log10(min(x)), np.log10(max(x)), 31)

# histogram 
log_counts, log_bin_edges = np.histogram(x, bins=log_bins)

# midpoints of the logarithmic bins
log_bin_midpoints = (log_bin_edges[:-1] + log_bin_edges[1:]) / 2

# Normalize the counts by the bin widths
log_bin_widths = np.diff(log_bin_edges)
normalized_log_counts = log_counts / log_bin_widths

plt.figure(figsize=(7, 6))

# Plot 
plt.loglog(log_bin_midpoints, normalized_log_counts, 'o', label='Logarithmically Binned Midpoints')

# reference line 
scaling_factor_log = 500
y_vals_log_binned = scaling_factor_log * x_vals_log**(-3/2)
plt.loglog(x_vals_log, y_vals_log_binned, 'r-', label='Cx^(-3/2)')


plt.title('Midpoint Plot with Logarithmic Binning (Log-Log Scale)')
plt.xlabel('x (bin midpoint)')
plt.ylabel('Normalized Frequency')
plt.legend()
plt.grid(True, which="both", ls="--")
plt.show()
```

e) (w = 1) (The advantage of logarithmic over linear coordinates is obvious. It
should also be obvious that midpoint plots are superior to histograms for
discerning functional forms.) Looking at your binned plots in c) and d), what did
you gain by doing logarithmic instead of linear binning? (2 ‚Äì 3 advantages
expected.)


**Answer**:

Advantages of using logarithmic binning:

**More Even Distribution of Data Points**: In the linear binning plot, most of the data points fall into the first few bins, leaving the rest of the bins with very few or no points. Logarithmic binning creates wider bins for larger x values, which helps to capture more of the sparse data in the tail of the distribution, leading to a more even number of data points per bin.

**Reduced Statistical Fluctuations in the Tail**: Because the logarithmic bins in the tail are wider, they average over a larger range of x values. This reduces the statistical noise and provides a more stable estimate of the pdf in the region where data is sparse.

**Clearer Visualization of the Power-Law Behavior**: With logarithmic binning, the log-log plot shows a clear linear trend, making the power-law behaviour and scaling factor (reference line) easier to identify. In contrast, linear binning produces a noisier tail that makes it harder to identify the trend.



f) (w < 1) One could try to simulate the pdf by rejection method. (You don‚Äôt need to
do the simulation.) Would this be slower or faster than the inverse distribution
method used here? Justify your answer.

**Answer**:
For power-law distributions, the rejection method is much less efficient than the inverse distribution method. Since the pdf has a long tail, any simple proposal distribution (e.g., a uniform rectangle) covers a far larger area than the target pdf. This makes the acceptance ratio very small, leading to frequent rejections and slow sampling. By contrast, the inverse method produces valid samples directly from each random draw.
