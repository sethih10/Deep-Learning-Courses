### Inhomogenous Poisson process

The arrival rate - $\lambda(t)$ 

```python
# value of lambda based on time interval 
def lambda_t(t): 
    '''
    Returns the arival rate at a give time t (in minutes)
    '''
    
    hour = t/60
    
    if 0 <= hour < 6:
        return 0.2
    elif 6 <= hour < 8:
        return 0.9
    elif 8 <= hour < 12:
        return 0.5
    elif 12 <= hour < 15:
        return 0.6
    elif 15 <= hour < 17:
        return 1.0
    elif 17 <= hour < 19:
        return 0.6
    else:
        return 0.5
```

### 1.a. Revised method

The revised method simplifies the inhomogenous poisson process into multiple independent homogenous process with constant $\lambda$ for each interval. 


From this logic, for a time interval [t1, t2] with a constant rate $\lambda$, the events in this interval follows a poisson distribution with mean $\lambda*(t2-t1)$


```python
def simulate_revised(T, lambda_max):
    
    events = []
    x_prev = 0.0
    
    # Sample number of events
    m = np.random.poisson(lambda_max*T)
    
    for i in range(1, m+1):
        
        u = np.random.rand()
        
        
        # generate ordered time
        x_i = x_prev + (T-x_prev)*(1 - u **(1/(m-i+1)))
        
        y = np.random.rand()
        
        if y<=lambda_t(x_i): 
            events.append(x_i)
        x_prev = x_i
        
    return np.array(events)

```

```python
np.random.seed(42)
T = 24*60
lambda_max = 1.0 # max rate from the table 

N_sim = 1000
t_grid = np.arange(0, T)
counts = np.zeros_like(t_grid, dtype=float)
cum_counts = np.zeros_like(t_grid, dtype=float)
first_arrivals = []
total_cars = []
```

```python
# running simulations 

for _ in range(N_sim): 
    arrivals = simulate_revised(T, lambda_max)
    
    if len(arrivals) == 0:
        continue
    total_cars.append(len(arrivals))
    first_arrivals.append(arrivals[0])
    hist, _ = np.histogram(arrivals, bins=np.arange(0, T+1))
    counts += hist
    cum_counts += np.cumsum(hist)
    
avg_counts = counts/N_sim
avg_cum = cum_counts/N_sim
avg_total = np.mean(total_cars)
avg_first = np.mean(first_arrivals)
```

(i) (w=2) For each such simulation, plot the average number of cars that arrived at (and obviously left) the crossing per minute over the period of 24 hours.


```python
plt.plot(t_grid/60, avg_counts)
plt.title('Average number of cars per minute (revised method)')
plt.xlabel('Time (hours)')
plt.ylabel('Average number of cars')
```

```python
plt.plot(t_grid/60, avg_cum)
plt.title('Cumulative number of cars per minute (revised method)')
plt.xlabel('Time (hours)')
plt.ylabel('Cumulative number of cars')
```

### 1.b. Thinning approach

In thinning approach, we sample the inter-event times from an $exp(U_{\lambda}$ distribution to get the x-coordinates (time), then sample the corresponding y-coordinates(number of cars) from a U(0, $U_{\lambda}$) and reject the points outside the envelope

```python
def simulate_thinning(T, lambda_max):
    
    event_times = []
    t = 0
    
    while t < T: 
        
        inter_arrival_time = np.random.exponential(lambda_max)
        
        t += inter_arrival_time
        
        if t>=T:
            break
            
            
        u = np.random.uniform(0,1)
        
        if u <= (lambda_t(t)/lambda_max):
            event_times.append(t)
        
    return np.array(event_times)
    
```

```python
np.random.seed(42)
T = 24*60
lambda_max = 1.0 # max rate from the table 

N_sim = 1000
t_grid = np.arange(0, T)
counts = np.zeros_like(t_grid, dtype=float)
cum_counts = np.zeros_like(t_grid, dtype=float)
first_arrivals = []
total_cars = []
```

```python
# running simulations 

for _ in range(N_sim): 
    arrivals = simulate_thinning(T, lambda_max)
    
    if len(arrivals) == 0:
        continue
    total_cars.append(len(arrivals))
    first_arrivals.append(arrivals[0])
    hist, _ = np.histogram(arrivals, bins=np.arange(0, T+1))
    counts += hist
    cum_counts += np.cumsum(hist)
    
avg_counts = counts/N_sim
avg_cum = cum_counts/N_sim
avg_total = np.mean(total_cars)
avg_first = np.mean(first_arrivals)
```

(i) (w=2) For each such simulation, plot the average number of cars that arrived at (and obviously

```python
plt.plot(t_grid/60, avg_counts)
plt.title('Average number of cars per minute (revised method)')
plt.xlabel('Time (hours)')
plt.ylabel('Average number of cars')
```

## Exercise 2 

2. In Assignment 3 you implemented and simulated homogeneous and inhomogeneous Poisson
process of radioactive decay by applying inversion distribution method to inter-event time
distribution. Here you will use the Poisson distribution for the number of events. Implement both
processes by drawing samples from a Poisson distribution to obtain number of events in
appropriate time intervals. Use ‚àÜùë° = 0.01 as the magnitude of one time step. So, elapsed time =
steps √ó ‚àÜùë°. The number of nuclei is initially ùëÅ0 = 10 000. The decay (fission) rate for one nucleus
is $\lambda_c$ = 0.3 per second. (Note that this $\lambda_c$ value is different from what was used in Assignment 3.)


```python
N0 = 10000
lambda_c = 0.3
dt = 0.01
```

a) Consider an ensemble of ùëÅ0 = 10 000 nuclei, where the number of undecayed nuclei is
kept constant; in other words, every time a nucleus decays it is replaced by an
undecayed nucleus. The differential equation for the process is $\frac{dN}{dt} = ‚àí\lambda_c ùëÅ_0$.
Accordingly, the constant decay rate is $\lambda = \lambda_c N_0$. Draw samples from the Poisson
distribution by using numpy.random.poisson. This is the number of events in a time
interval.

### Homogenous process

Number of undecayed nuclei is kept constant. 

Decay rate is kept constant $\lambda = \lambda_c N_0$

Number of decays k in a small time interval $\Delta t$

$ k \sim Poisson(\lambda(t) \Delta t) $



(i) (w=2) Implement the algorithm. Indicate clearly the different parts and functions.


```python
def homogenous_decay(N0, lambda_c, dt):
    
    lambda_t = lambda_c*N0 # constant decay rate
    
     
    step = 0
    decayed = 0
    
    # simulate for half time (until half of the nuclei is left)
    while decayed < N0/2:
        
        k = np.random.poisson(lambda_t*dt)
        
        decayed += k
        step += 1
        
    return dt*step
    
    
```


(ii) (w=2) Compute mean values by averaging over 1000 simulations of the following:

The half time, i.e., the time it takes for $ùëÅ_{0}$/2 nuclei to decay, and the variance of this
half time.


```python
num_simulations = 1000
half_times_homogenous = []

# Run the simulation for 1000 times to calculate the mean half time 
for _ in range(num_simulations): 
    half_time = homogenous_decay(N0, lambda_c, dt)
    half_times_homogenous.append(half_time)
    

mean_half_time_homo = np.mean(half_times_homogenous)
var_half_time_homo = np.var(half_times_homogenous)

print("Homogenous Poisson process:")
print(f"Mean half-time: {mean_half_time_homo:.2f} s")
print(f"Variance of half-time: {var_half_time_homo:.5f} $s^2$")
```

b) Next, consider an ensemble of ùëÅ0 = 10 000 nuclei, where decayed nuclei are not
replaced. The process becomes inhomogeneous as the decay rate will depend on time.
The differential equation describing this process reads as 
$\frac{dN}{dt} = ‚àí\lambda_c ùëÅ(ùë°)$, where the number of undecayed nuclei decreases exponentially, $ùëÅ(ùë°) = ùëÅ_0 e^{-\lambda_c t}$. The inter-event times are different from the homogeneous case in a). Accordingly, you need to sample number of events in appropriate time interval using numpy.random.poisson.


### Inhomogenous process 

Undecayed nuclei are not constant

$ùëÅ(ùë°) = ùëÅ_0 e^{-\lambda_c t}$

Rate of decay is not constant
$ \lambda (t) = \lambda_c N_0 e^{-\lambda_c t}$



(i) (w=2) Implement the algorithm. Indicate clearly the different parts and functions.



```python
def inhomogenous_decay(N0, lambda_c, dt): 
    
    total_time = 0.0
    
    decayed_nuclei = 0
    steps = 0
    
    while decayed_nuclei < N0/2:
        
        lambda_t = lambda_c*(N0 - decayed_nuclei)
        
        decay = np.random.poisson(lambda_t*dt)
        
        decayed_nuclei += decay
        
        total_time += dt
        
    return total_time
        
```

(ii) (w=2) Compute mean values by averaging over 1000 simulations of the following:

The half time, i.e., the time it takes for ùëÅ0/2 nuclei to decay, and the variance of this
half time.


Note. Above, ùëÅ = „ÄàùëÅ„Äâ, i.e. an ensemble average. Likewise, ùëÅ(ùë°) = „ÄàùëÅ(ùë°)„Äâ, i.e. an ensemble
average at time ùë°.

```python
# we run this simulation 1000 times to calculate the mean half time

num_simulations = 1000
half_times_inhomogenous = []
for _ in range(num_simulations): 
    
    half_time = inhomogenous_decay(N0, lambda_c, dt)
    half_times_inhomogenous.append(half_time)
    
mean_in = np.mean(half_times_inhomogenous)
variance_in = np.var(half_times_inhomogenous)

print("Inhomogeneous Poisson Process:")
print(f"Mean half-time: {mean_in:.2f} s")
print(f"Variance of half-time: {variance_in:.5f} $s^2$")
```
